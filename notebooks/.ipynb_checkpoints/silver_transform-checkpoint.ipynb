```python
import os
import pyspark
from pyspark.sql import SparkSession
from pyspark.sql.functions import current_timestamp

os.environ["AWS_REGION"] = "us-east-1"
os.environ["AWS_DEFAULT_REGION"] = "us-east-1"
os.environ["AWS_ACCESS_KEY_ID"] = "admin"
os.environ["AWS_SECRET_ACCESS_KEY"] = "password"

CATALOG_URI = "http://nessie:19120/api/v1"
WAREHOUSE = "s3a://warehouse/"
S3_ENDPOINT = "http://minio:9000"
AWS_KEY = "admin"
AWS_SECRET = "password" 
```

```python
# ── Crear SparkSession ──
conf = (
    pyspark.SparkConf()
    .setAppName("silver_notebook")
    .set("spark.driver.memory", "2g")
    .set("spark.sql.shuffle.partitions", "4")
    .set("spark.sql.adaptive.enabled", "true")
    .set("spark.sql.adaptive.coalescePartitions.enabled", "true")
    .set("spark.jars.packages",
        "org.apache.hadoop:hadoop-aws:3.3.4,"
        "com.amazonaws:aws-java-sdk-bundle:1.12.262,"
        "software.amazon.awssdk:bundle:2.24.8,"
        "software.amazon.awssdk:url-connection-client:2.24.8,"
        "org.apache.iceberg:iceberg-spark-runtime-3.5_2.12:1.5.0,"
        "org.projectnessie.nessie-integrations:nessie-spark-extensions-3.5_2.12:0.77.1")
    .set("spark.sql.extensions",
        "org.apache.iceberg.spark.extensions.IcebergSparkSessionExtensions,"
        "org.projectnessie.spark.extensions.NessieSparkSessionExtensions")
    .set("spark.sql.catalog.nessie", "org.apache.iceberg.spark.SparkCatalog")
    .set("spark.sql.catalog.nessie.uri", CATALOG_URI)
    .set("spark.sql.catalog.nessie.ref", "main")
    .set("spark.sql.catalog.nessie.authentication.type", "NONE")
    .set("spark.sql.catalog.nessie.catalog-impl", "org.apache.iceberg.nessie.NessieCatalog")
    .set("spark.sql.catalog.nessie.warehouse", WAREHOUSE)
    .set("spark.sql.catalog.nessie.io-impl", "org.apache.iceberg.aws.s3.S3FileIO")
    .set("spark.sql.catalog.nessie.s3.endpoint", S3_ENDPOINT)
    .set("spark.sql.catalog.nessie.s3.path-style-access", "true")
    .set("spark.sql.catalog.nessie.s3.access-key-id", AWS_KEY)
    .set("spark.sql.catalog.nessie.s3.secret-access-key", AWS_SECRET)
    .set("spark.sql.catalog.nessie.s3.region", "us-east-1")
    .set("spark.hadoop.fs.s3a.endpoint", S3_ENDPOINT)
    .set("spark.hadoop.fs.s3a.access.key", AWS_KEY)
    .set("spark.hadoop.fs.s3a.secret.key", AWS_SECRET)
    .set("spark.hadoop.fs.s3a.path.style.access", "true")
    .set("spark.hadoop.fs.s3a.impl", "org.apache.hadoop.fs.s3a.S3AFileSystem")
    .set("spark.hadoop.fs.s3a.aws.credentials.provider", "org.apache.hadoop.fs.s3a.SimpleAWSCredentialsProvider")
    .set("spark.hadoop.fs.s3a.connection.ssl.enabled", "false")
    .set("spark.hadoop.fs.s3a.endpoint.region", "us-east-1")
    .set("spark.driver.extraJavaOptions",
         "-Daws.region=us-east-1 -Daws.accessKeyId=admin -Daws.secretAccessKey=password")
)

spark = SparkSession.builder.config(conf=conf).getOrCreate()
spark.sparkContext.setLogLevel("WARN")
print(f"Spark: {spark.version}")
```

```python
# ── Funciones auxiliares ──
def table_exists(name):
    try:
        spark.read.format("iceberg").load(name).limit(1).collect()
        return True
    except:
        return False

def merge_manual(df_new, table_name, key_col="Id"):
    if not table_exists(table_name):
        print(f"  Creando {table_name}...")
        df_new.writeTo(table_name).using("iceberg").tableProperty("format-version", "2").create()
        return
    print(f"  Merge manual en {table_name}...")
    df_existing = spark.read.format("iceberg").load(table_name)
    df_unchanged = df_existing.join(df_new.select(key_col), on=key_col, how="left_anti")
    df_merged = df_unchanged.unionByName(df_new, allowMissingColumns=True)
    df_merged.writeTo(table_name).using("iceberg").overwritePartitions()
    print(f"  Merge completado.")
```

```python
# ── Namespace ──
spark.sql("CREATE NAMESPACE IF NOT EXISTS nessie.silver")
```

```python
# ── Silver: users_hist ──
print("Leyendo users desde Bronze...")
df_u2020 = spark.read.parquet("s3a://bronze/users/2020/users_2020.parquet")
df_u2021 = spark.read.parquet("s3a://bronze/users/2021/users_2021.parquet")
df_users = df_u2020.unionByName(df_u2021, allowMissingColumns=True)
df_users = df_users.withColumn("fecha_cargue", current_timestamp())
print(f"Users combinados: {df_users.count()}")

merge_manual(df_users, "nessie.silver.users_hist", key_col="Id")
print("Silver users completado.")
```

```python
# ── Silver: posts_hist (sin Body) ──
COLS = ["Id", "PostTypeId", "AcceptedAnswerId", "CreationDate",
        "Score", "ViewCount", "OwnerUserId", "LastActivityDate",
        "Title", "Tags", "AnswerCount", "CommentCount",
        "FavoriteCount", "ContentLicense", "ParentId"]

print("Leyendo posts desde Bronze (sin Body)...")
df_p2020 = spark.read.parquet("s3a://bronze/posts/2020/posts_2020.parquet").select(*COLS)
df_p2021 = spark.read.parquet("s3a://bronze/posts/2021/posts_2021.parquet").select(*COLS)
df_posts = df_p2020.unionByName(df_p2021, allowMissingColumns=True)
df_posts = df_posts.withColumn("fecha_cargue", current_timestamp())
print(f"Posts combinados: {df_posts.count()}")

df_posts.writeTo("nessie.silver.posts_hist").using("iceberg").createOrReplace()
print("Silver posts completado.")
```

```python
# ── Verificar ──
print("=" * 50)
print("RESUMEN SILVER")
print("=" * 50)
for t in ["nessie.silver.users_hist", "nessie.silver.posts_hist"]:
    try:
        cnt = spark.read.format("iceberg").load(t).count()
        print(f"  {t}: {cnt} registros")
    except Exception as e:
        print(f"  {t}: ERROR - {e}")
print("=" * 50)
spark.stop()
```
